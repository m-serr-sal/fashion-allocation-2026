```python
import pandas as pd
import numpy as np
```

```python
products = pd.DataFrame({
    "product_id": range(1, 21),
    "category": np.random.choice(
        ["Dress","Blazer","Cargo Pants","Knit Top","Denim"],
        20
    ),
    "price": np.random.randint(80, 400, 20),
    "silhouette": np.random.choice(
        ["Tailored","Relaxed","Oversized","Fitted"],
        20
    ),
    "fabric": np.random.choice(
        ["Silk","Wool","Cotton","Nylon","Denim"],
        20
    ),
    "trend_tags": np.random.choice(
        ["Quiet Luxury","Utility","Techwear","Coastal Chic","Power Dressing"],
        20
    )
})
```

```python
products.to_csv("../data/products.csv", index=False)
```
