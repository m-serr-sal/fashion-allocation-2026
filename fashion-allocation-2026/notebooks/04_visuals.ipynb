```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

allocations = pd.read_csv("../data/allocations.csv")

allocations.head()
```

```python
pivot = allocations.pivot(
    index="product_id",
    columns="city",
    values="score"
)
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    pivot,
    cmap="Blues",
    annot=True
)

plt.title("Product-Store Allocation Scores")
plt.show()
```
